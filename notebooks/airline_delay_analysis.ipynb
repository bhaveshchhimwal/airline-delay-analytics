```python
df = flights[
    [
        "YEAR",
        "MONTH",
        "DAY",
        "DAY_OF_WEEK",

        "AIRLINE",
        "FLIGHT_NUMBER",

        "ORIGIN_AIRPORT",
        "DESTINATION_AIRPORT",

        "SCHEDULED_DEPARTURE",
        "DEPARTURE_TIME",
        "DEPARTURE_DELAY",

        "SCHEDULED_ARRIVAL",
        "ARRIVAL_TIME",
        "ARRIVAL_DELAY",

        "DISTANCE",
        "CANCELLED",
        "CANCELLATION_REASON"
    ]
]
```

```python
df = df[df["CANCELLED"] == 0]
```

```python
df = df.drop(
    columns=["CANCELLATION_REASON"]
)
```

```python
df.reset_index(
    drop=True,
    inplace=True
)
```

```python
df = df.dropna()

df.reset_index(
    drop=True,
    inplace=True
)
```

```python
df.shape
```

```python
df = df.merge(
    airlines,
    left_on="AIRLINE",
    right_on="IATA_CODE",
    how="left"
)

df.head()
```

```python
df.drop(
    columns=[
        "AIRLINE_x",
        "IATA_CODE"
    ],
    inplace=True
)
```

```python
df.rename(
    columns={
        "AIRLINE_y":"AIRLINE"
    },
    inplace=True
)
```

```python
df.head()
```

```python
df["DATE"] = pd.to_datetime(
    df[["YEAR", "MONTH", "DAY"]]
)
```

```python
df[
    [
        "YEAR",
        "MONTH",
        "DAY",
        "DATE"
    ]
].head()
```

```python
df["MONTH_NAME"] = df["DATE"].dt.month_name()
```

```python
df[
    [
        "DATE",
        "MONTH",
        "MONTH_NAME"
    ]
].head()
```

```python
df["IS_DELAYED"] = np.where(
    df["ARRIVAL_DELAY"] > 15,
    1,
    0
)
```

```python
df["IS_DELAYED"].value_counts()
```

```python
delay_percentage = (
    df["IS_DELAYED"].mean() * 100
)

delay_percentage
```

```python
def delay_status(delay):

    if delay <= 0:
        return "Early / On Time"

    elif delay <= 15:
        return "Slight Delay"

    elif delay <= 60:
        return "Moderate Delay"

    else:
        return "Major Delay"


df["DELAY_STATUS"] = (
    df["ARRIVAL_DELAY"]
    .apply(delay_status)
)
```

```python
df["DELAY_STATUS"].value_counts()
```

```python
def time_period(time):

    hour = int(time // 100)

    if 5 <= hour < 12:
        return "Morning"

    elif 12 <= hour < 17:
        return "Afternoon"

    elif 17 <= hour < 21:
        return "Evening"

    else:
        return "Night"


df["TIME_PERIOD"] = (
    df["SCHEDULED_DEPARTURE"]
    .apply(time_period)
)
```

```python
df["TIME_PERIOD"].value_counts()
```
